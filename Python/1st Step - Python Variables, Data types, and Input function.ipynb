## Performing Arithmetic Operations Using Python
Python supports basic arithmetic operations like:
- Addition (`+`)
- Subtraction (`-`)
- Multiplication (`*`)
- Division (`/`)
- Exponentiation (`**`)
- Modulus (`%`)
- Integer Division (`//`)


```python
# Examples of arithmetic operations
a = 10
b = 3
print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Exponentiation:", a ** b)
print("Modulus:", a % b)
print("Integer Division:", a // b)
```

### Advanced Arithmetic Operations
Python supports more complex arithmetic calculations:

```python
 (5 - 1) * ((7 + 1) / (3 - 1))
```

### `type()` Function

The `type()` function in Python is used to determine the data type of a variable or value. It returns the type of the object passed as its argument.

`type(object)`

```python
x = 10
print(type(x))  

y = 3.14
print(type(y))  

z = "Hello"
print(type(z))  
```

### `input()` Function in Python

The `input()` function is used to take input from the user in Python. It allows the user to enter a value, which is returned as a string by default.

```python
name = input("Enter your name: ")
print("Hello, " + name + "!")
```

##### Converting Input to Other Data Types

Since `input()` returns a string, you may need to convert the input into other data types (like integer, float, etc.) for calculations.

```python
age = int(input("Enter your age: "))
print("In 5 years, you will be:", age + 5)
```

## Evaluating Conditions Using Python
Python provides comparison operators to evaluate conditions:
- `==`: Equal to
- `!=`: Not equal to
- `<`: Less than
- `>`: Greater than
- `<=`: Less than or equal to
- `>=`: Greater than or equal to


```python
# Example of evaluating conditions
x = 10
y = 20
print("Is x equal to y?", x == y)
print("Is x less than y?", x < y)
print("Is x greater than or equal to y?", x >= y)
```

## Combining Conditions with Logical Operators
Python allows combining multiple conditions using:
- `and`: Both conditions must be true
- `or`: At least one condition must be true
- `not`: Negates the condition


```python
# Example of combining conditions
x = 15
print("Is x between 10 and 20?", x > 10 and x < 20)
print("Is x less than 10 or greater than 20?", x < 10 or x > 20)
print("Is x not equal to 15?", not (x == 15))
```

## Adding Text Styles Using Markdown
You can style text in Markdown using:
- **Headings:** `#`, `##`, `###` for levels 1, 2, 3

- *Italics:* `*text*`
- `Code`: `` `code` ``
- **Lists:**
  - Unordered: `-` or `*`
  - Ordered: `1.`, `2.`
- **Hyperlinks:** `[link text](URL)`
- **Images:** `![alt text](image_url)`
